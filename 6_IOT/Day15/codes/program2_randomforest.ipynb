## Ensemble Learning

#### predict whether a custom will exist the bank

```python
# pre-requisites

import warnings
warnings.filterwarnings(action="ignore")
```

```python
# import required packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# load the datset
df=pd.read_csv('Churn_Modelling.csv')
df.head()
```

```python
df.info()
```

```python
df.isna().sum()
```

### Data Cleansing

```python
df.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
```

```python
from sklearn.preprocessing import LabelEncoder

df['Geography'] = LabelEncoder().fit_transform(df['Geography'])
df['Gender'] = LabelEncoder().fit_transform(df['Gender'])
```

```python
df.corr()
```

```python
df.drop(['EstimatedSalary'],axis=1,inplace=True)
```

#### split the data

```python
# create x
x=df.drop('Exited',axis=1)

# create y
y=df['Exited']
```

```python
from sklearn.model_selection import train_test_split

# split into train and test sets
x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=123456)
```

### Model Building

```python
from sklearn.tree import DecisionTreeClassifier

# create the model
model_decision = DecisionTreeClassifier(criterion='gini',max_depth=1000)

# train the model
model_decision.fit(x_train,y_train)
```

```python
y_pred = model_decision.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

accuracy =accuracy_score(y_test,y_pred)
precision =precision_score(y_test,y_pred)
recall =recall_score(y_test,y_pred)
f1 =f1_score(y_test,y_pred)

print(f"accuracy = {accuracy:.2f}")
print(f"Precision = {precision:.2f}")
print(f"recall = {recall:.2f}")
print(f"f1 score = {f1:.2f}")
```

#### Build the Random Forest Model

```python
from sklearn.ensemble import RandomForestClassifier

# create the model
model_random = RandomForestClassifier(n_estimators=60,max_depth=1000)

# train the model
model_random.fit(x_train,y_train)
```

```python
y_pred = model_random.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

accuracy =accuracy_score(y_test,y_pred)
precision =precision_score(y_test,y_pred)
recall =recall_score(y_test,y_pred)
f1 =f1_score(y_test,y_pred)

print(f"accuracy = {accuracy:.2f}")
print(f"Precision = {precision:.2f}")
print(f"recall = {recall:.2f}")
print(f"f1 score = {f1:.2f}")
```
