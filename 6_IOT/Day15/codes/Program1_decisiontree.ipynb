## Split the data

```python
# create x
x=df[['sex','cp','thalach','exang','oldpeak','slope', 'ca', 'thal']]

# create y
y=df['target']
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=123456)
```

### Build the model

```python
from sklearn.tree import DecisionTreeClassifier

# create the model
model = DecisionTreeClassifier(criterion='gini',max_depth=500)

# train the model
model.fit(x_train,y_train)
```

```python
# grid search
```

### model evaluation

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,precision_score,recall_score,f1_score

accuracy =accuracy_score(y_test,y_pred)
precision =precision_score(y_test,y_pred)
recall =recall_score(y_test,y_pred)
f1 =f1_score(y_test,y_pred)

print(f"accuracy = {accuracy:.2f}")
print(f"Precision = {precision:.2f}")
print(f"recall = {recall:.2f}")
print(f"f1 score = {f1:.2f}")
```

### visualization

```python
plt.scatter(x_test['thalach'][y_pred==0],x_test['oldpeak'][y_pred==0],label="won't suffer",color="green")
plt.scatter(x_test['thalach'][y_pred==1],x_test['oldpeak'][y_pred==1],label="Will suffer",color="red")
plt.xlabel("thalach")
plt.ylabel("oldpeak")
plt.title("decision tree")
plt.legend()
```

```python
# tree visualization
from sklearn.tree import plot_tree

tree = plot_tree(model,max_depth=1,filled=True)
```
