### Create the Pandas series using dictionary

```python
s1=pd.Series({"Name":"Sujata","org":"Sunbeam","Course":"Python"})
print(s1)
```

```python
students =pd.Series(["s1","s2","s3","s4","s5"])
marks = pd.Series([35,20,25,30,40])
print(students)
print(marks)
```

```python
marks = pd.Series([35,20,25,30,40])
students =pd.Series(["s1","s2","s3","s4","s5"],index=marks)
print(students)
```

### Create aa dataframe using dictionary

```python
data = {'Name':['n1','n2','n3','n4'],'Age':[23,22,25,24]}
df = pd.DataFrame(data,index=['rank1','rank2','rank3','rank4'])
print(df)
```

```python
def function1():
    p1=({"name":"P1","age":25,"id": 1234})
    p2=({"name":"P2","age":23,"id": 5678})
    p3=({"name":"P3","age":24,"id": 9876})

    s1 =pd.Series(p1)
    s2 =pd.Series(p2)
    s3 =pd.Series(p3)

    df=pd.DataFrame([s1,s2,s3])
    print(s1)
    print(s2)
    print(s3)
```

```python
function1()
```

```python
def function2():
    p1=({"name":"P1","age":25,"id": 1234})
    p2=({"name":"P2","age":23,"id": 5678})
    p3=({"name":"P3","age":24,"marks":85})

    s1 =pd.Series(p1)
    s2 =pd.Series(p2)
    s3 =pd.Series(p3)

    df=pd.DataFrame([s1,s2,s3])
    print(df)
```

```python
function2()
```
