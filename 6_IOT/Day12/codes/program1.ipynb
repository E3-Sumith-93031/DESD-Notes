### numpy array creation example, array() and print the properties

```python
import numpy as np
```

```python
def function1():
    arr1 = [10,20,30,40,50]
    print(arr1)
    print(f"Type = {type(arr1)}")
    print(f"arr1[0]Type = {type(arr1[0])}")
```

```python
  function1()
```

```python
def function2():
    n_array = np.array([11,22,33,44,55])
    print(n_array)
    print(f"Type = {type(n_array)}")
```

```python
function2()
```

```python
def function3():
    arr = np.array([10,20,30,40,50,60])
    print(arr)
    print(f"Type = {type(arr)}")
    print(f"DType = {arr.dtype}")
    print(f"Item size = {arr.itemsize}")
    print(f"Size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")
```

```python
function3()
```

```python
def printProperties(arr):
    print(arr)
    print(f"Type = {type(arr)}")
    print(f"DType = {arr.dtype}")
    print(f"Item size = {arr.itemsize}")
    print(f"Size = {arr.size}")
    print(f"NDIM = {arr.ndim}")
    print(f"NBYTES = {arr.nbytes}")
n1 = np.array([2,5,8,9,3,4])
printProperties(n1)
```

```python
n3 = np.array([11,22,33,44],dtype = np.int32)
printProperties(n3)
```

```python
n4= np.array([1.15,2.31,3.8])
printProperties(n4)
```

```python
n5 = np.array(["C","C++","Java","Python"])
printProperties(n5)
```

```python
country=np.array(["India","USA","Australia"])
printProperties(country)
```

## Indexing,Slicing,reshape

```python
# if you are making a new file
# import numpy as np
```

```python
n1 = np.array([11,22,33,44,55,66,77,88])
print(n1)
```

```python
print(n1[2])    # positive indexing
```

```python
print(len(n1))   # no of elements in the array
```

```python
print(n1[4])      # positive indexing
```

```python
print(n1[-1])     # negative indexing
```

```python
print(n1[-8])     # negative indexing
```

```python
print(n1[0])     # positive indexing
```

```python
print(n1[2:5])    # start[2]= slicing starts      stop[5]=slicing will end at 5(but the last index is not included in the output)
```

```python
print(f"NDIM ={n1.ndim}" )
```

```python
print(f"Shape = {n1.shape}")
```

```python
n1= n1.reshape([4,2])
print(n1)
```

```python
print(f"NDIM ={n1.ndim}" )
```

```python
print(f"Shape = {n1.shape}")
```

```python
n1=np.array([11,22,33,44])
n2=np.array([10,20,30,40])
print(f"N1 = {n1}")
print(f"N2 = {n2}")
print(n1+n2)
```

```python
n1=np.random.random(5)      # generate any random 5 floating-point numbers 
print(n1)                   # each number is between 0.0 to 1.0(1.0 not included)
```

```python
n4=np.random.randint(1,10,4)       # any four integer numbers are randomly generated between 1 to 10       
print(n4)   
```
