## Data Cleansing Process

#### Dela with NaN values or missing values
- missing value in the data set
- if values are missing, the data set can not be used for further processing 
- methods
- remove the row(s) having NaN values
- remove the column(s) having NaN values
- replace the NaN values with mean, mode or median

```python
# import required packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# axis = 0 for rows
# dropna() removes the row(s) containing at least one NA records and
# returns a new data frame (the original DF does NOT get affected)
df_new = df.dropna(axis=0)
df_new
```

```python
# instead of returning a new df, modify or update the existing df
# df.dropna(axis=0, inplace=True)
```

```python
# axis = 1 for columns
# dropna() removes the column(s) containing at least one NA record and
# returns a new data frame (the original DF does NOT get affected)
df_new = df.dropna(axis=1)
df_new
```

```python
# instead of returning a new df, modify or update the existing df
# df.dropna(axis=1, inplace=True)
```

### replace the NA values with mean, mode or median

```python
df['age'].mean()
np.float64(29.8811345124283)
# df['age'].fillna(df['age'].mean(), inplace=True)
df['age'] = df['age'].fillna(df['age'].mean())
```

#### replace the NA values with Mode of respective columns

```python
cabin_mode = df['cabin'].mode()[0]
df['cabin'].mode()[0]
```

```python
df['cabin'] = df['cabin'].fillna(cabin_mode)
```
