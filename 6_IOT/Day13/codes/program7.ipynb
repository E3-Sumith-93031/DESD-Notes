## Simple Linear Regression

#### Salary Prediction
- predict the salary on basis of yrs of experience of an employee

```python
# import required packages

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# prequisities

import warnings
warnings.filterwarnings(action='ignore')
```

```python
# load the dataset
df=pd.read_csv('Salary_Data.csv')
df.head()
```

### EDA
- get the basic information about the dataset

```python
df.info()
```

```python
# get the statistical informaation about the dataset
df.describe()
```

```python
# to check for NA value
df.isna().sum()
```

### split the data

```python
# create x
x=df.drop('Salary',axis=1)

# creat y
y=df['Salary']
```

```python
df.head()
```

```python
# split the data as train and test

from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,train_size=0.8,random_state=123456)
```

### Model Building

```python
from sklearn.linear_model import LinearRegression

# create the model
model = LinearRegression()

# train the model using x and y
model.fit(x,y)
```

### Model Prediction

```python
prediction = model.predict([[16]])
print(f"Salary of an employee after 16 years of experience = {prediction[0]:.2f}")
```

```python
# predict the salary of an employee 15,17,100 years of experience
```

```python
y_pred = model.predict(x_test) 
```

### Evaluation Metrics

```python
from sklearn.metrics import mean_absolute_error

mae = mean_absolute_error(y_test,y_pred)
mae
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_test,y_pred)
mse
```

```python
from sklearn.metrics import root_mean_squared_error

rmse = root_mean_squared_error(y_test,y_pred)
rmse
```

```python
from sklearn.metrics import r2_score

r2 = r2_score(y_test,y_pred)
r2
```
